## Group the Data by Day of Week and Plot the Sum of The Numeric Columns

```python
#Your code here
df.columns
type(df["DATE"])  # pandas.core.series.Series

df.DATE = pd.to_datetime(df.DATE) #Convert to datetime column

df["Day_of_Week"] = df.DATE.dt.dayofweek 
print(df["Day_of_Week"].tail())

grouped = df.groupby("Day_of_Week").sum()
grouped.plot(kind = "bar") # Or, kind = "barh"  -- Horizontal Bar
```

## Group the Data by Weekend/Weekday and Plot the Sum of the Numeric Columns

```python
#Your code here
#grouped = grouped.reset_index()
grouped.head()

grouped["Is_WeekEnd"] = grouped.Day_of_Week.map({0:False,1:False,2:False,3:False,4:False,5:False,6:True,7:True})
wkend = grouped.groupby("Is_WeekEnd").mean()
wkend[["ENTRIES","EXITS"]].plot(kind = "bar")
```

```python
# Do not use SUM to solve this problem statement ....
grouped["Is_WeekEnd"] = grouped.Day_of_Week.map({0:False,1:False,2:False,3:False,4:False,5:False,6:True,7:True})
wkend = grouped.groupby("Is_WeekEnd").sum()
wkend[["ENTRIES","EXITS"]].plot(kind = "bar")
```
